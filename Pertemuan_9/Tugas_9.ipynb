```python
print("Diberikan data list berikut:")
data_buah = ["apel","mangga","jeruk","anggur","pisang"]
print(data_buah)

print("\na. Menambahkan 'semangka' di akhir list")
data_buah.append("semangka")
print(f"Data setelah ditambah: {data_buah}")

print("\nb. Menyisipkan 'durian' di antara 'jeruk' dan 'anggur'")
data_buah.insert(3,"durian")
print(print(f"Data setelah ditambah: {data_buah}"))

print("\nc. Menghapus 'mangga' dari list")
data_buah.remove("mangga")
print(print(f"Data setelah dihapus: {data_buah}"))

print("\nd. Mengubah 'pisang' menjadi 'nanas'")
data_buah[4] = "nanas"
print(print(f"Data setelah diubah: {data_buah}"))

print("\ne. Menampilkan 3 buah pertama")
data = data_buah[:3]
print(f"Tiga buah pertama: {data}")



```

```python
print("Diberikan list berikut:")
data_angka = [45,12,78,23,56,89,34]
print(data_angka)

print("\na. Urutkan secara ascending")
data_angka.sort()
print(f"Data angka ascending = {data_angka}")

print("\nb. Urutkan secara descending")
data_angka.reverse()
print(f"Data angka descending = {data_angka}")
```
