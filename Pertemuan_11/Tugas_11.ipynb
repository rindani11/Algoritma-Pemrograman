```python
import datetime
import random
import string

# Template data buku
buku_template = {
    'judul': 'Judul Buku',
    'penulis': 'Penulis',
    'tahun_terbit': 0,
    'tanggal_terakhir_pinjam': datetime.datetime(1111, 1, 1),
    'status': 'Tersedia'
}

# Dictionary utama untuk menyimpan semua buku
data_buku = {}

while True:
    print(f"{'SISTEM MANAJEMEN PERPUSTAKAAN':^50}")
    print("-" * 50)

    # Membuat dictionary buku baru dari template
    buku = dict.fromkeys(buku_template.keys())

    buku['judul'] = input("Judul Buku\t\t: ")
    buku['penulis'] = input("Penulis\t\t\t: ")
    buku['tahun_terbit'] = int(input("Tahun Terbit\t\t: "))

    # Input tanggal terakhir dipinjam
    tahun = int(input("Tahun terakhir pinjam (YYYY)\t: "))
    bulan = int(input("Bulan (1-12)\t\t: "))
    tanggal = int(input("Tanggal (1-31)\t\t: "))
    buku['tanggal_terakhir_pinjam'] = datetime.datetime(tahun, bulan, tanggal)

    buku['status'] = input("Status (Tersedia/Tidak)\t: ")

    KEY = ''.join(random.choice(string.ascii_uppercase) for _ in range(6))
    data_buku.update({KEY: buku})

    # Menampilkan seluruh data buku
    print("\nDaftar Buku:")
    print(f"{'KEY':<8} {'Judul':<25} {'Penulis':<15} {'Tahun':<6} {'Terakhir Pinjam':<15} {'Status':<10}")
    print("-" * 80)

    for key in data_buku:
        item = data_buku[key]
        print(f"{key:<8} {item['judul']:<25} {item['penulis']:<15} {item['tahun_terbit']:<6} "
              f"{item['tanggal_terakhir_pinjam'].strftime('%x'):<15} {item['status']:<10}")

    print("\n")
    lanjut = input("Tambah buku lagi? (y/n): ").lower()
    if lanjut == 'n':
        break

print("\nProgram selesai. Terima kasih!")
```
