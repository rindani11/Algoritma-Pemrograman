Dictionary

```python
keranjang ={
    'apel' : 5000,
    'pisang':3000,
    'jeruk': 7000,
    'anggur': 15000
}
total = 0
for buah in keranjang.values():
  total += buah

for barang in keranjang.items():
  print(f'Harga {barang}: Rp {buah}')
print(f" Total harga : Rp {total}")
```
