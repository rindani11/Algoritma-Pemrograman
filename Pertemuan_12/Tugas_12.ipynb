```python
print("Membuat Program dengan Input dan Fungsi untuk Menghitung")


# Luas persegi panjang
def luas_ppanjang(panjang, lebar):
  hasil = panjang*lebar
  return hasil

p = float(input("\n\nMasukan panjang persegi panjang "))
l = float(input("Masukan lebar persegi panjang "))
print("Luas Persegi Panjang = panjang x lebar")
print(f'Luas Persegi Panjang = {luas_ppanjang(p,l)}')

# Volume balok
def volume_balok(panjang, lebar, tinggi):
  hasil = panjang*lebar*tinggi
  return hasil

p = float(input("\nMasukan panjang balok "))
l = float(input("Masukan lebar balok "))
t = float(input("Masukan tinggi balok "))
print("Volume Balok = panjang x lebar x tinggi")
print(f'Volume Balok = {volume_balok(p,l,t)}')

# Luas lingkaran
def luas_lingkaran(jari):
  phi = 3.14
  return phi * jari * jari

j = float(input("\nMasukan jari-jari lingkarang "))
print("Luas Lingkarang = phi x jari x jari")
print(f"Luas Lingkaran = {luas_lingkaran(j):.3f}")

# Volume bola
def volume_bola(jari):
  phi = 3.14
  return (4/3) * phi * (jari ** 3)

jb = float(input("\nMasukan jari-jari bola "))
print("Volume Bola = 4/3 x phi x jari^3")
print(f"Volume Bola = {volume_bola(jb):.3f}")
```
